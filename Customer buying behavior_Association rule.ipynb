```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data=pd.read_csv('before-covid.csv')
data
```

```python
df=pd.DataFrame(data)
df
```

# we can generate frequent itemsets based on support.

```python
frequent_itemsets = apriori(df[['Baby Food','Fish','Frozen','Fruits','Grocery','Meat','Vegetables','Others']], min_support=0.5,use_colnames=True)
```

```python
frequent_itemsets
```

```python
frequent_itemsets.head(15)
```

```python
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
```

```python
rules
```

```python
rules.head(30)
```

```python
rules [ (rules['lift'] >1.1396)  & (rules['confidence']> 0.95)  ]
```
